Another technique to mitigate the exploding gradients problem is to clip the gradients during backpropagation so that they never exceed some threshold. This is called gradient clipping.‚Å† This technique is generally used in recurrent neural networks, where using batch normalization is tricky

During backpropagation, gradients can sometimes become extremely large, especially in:

* Recurrent Neural Networks (RNNs)
* Deep networks
* Networks with poor initialization
* High learning rates

Large gradients cause:
* Huge weight updates
* Loss becoming NaN
* Training crashing or oscillating

Gradient clipping prevents this.

Gradient clipping limits the magnitude of gradients during backpropagation.

If a gradient becomes too large, it is scaled down to a safe range.
